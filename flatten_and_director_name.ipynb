# flatten movie details

```python
import pandas as pd
import itertools
import ast
```

```python
movies_metadata_df = pd.read_csv('movies_metadata.csv')
```

```python
movie_names = movies_metadata_df['original_title']
production_companies = movies_metadata_df['production_companies']
production_countries = movies_metadata_df['production_countries']
genres = movies_metadata_df['genres']
```

```python
movies_all_combination_list = []
```

```python
from flatten_dict import flatten
```

```python
for i, movie_name in movie_names.items():
    try:
        movie_production_companies =  ast.literal_eval(movies_metadata_df.at[i, 'production_companies'])
        movie_production_countries =  ast.literal_eval(movies_metadata_df.at[i, 'production_countries'])
        movie_genres = ast.literal_eval(movies_metadata_df.at[i, 'genres'])
    except:
        pass
    
    for movie_production_company, movie_production_country, movie_genre in itertools.product(movie_production_companies, movie_production_countries, movie_genres):
        movie_details = {'movie_name': movie_name}
        movie_production_company_dict = ast.literal_eval(str(movie_production_company))
        movie_production_country_dict = ast.literal_eval(str(movie_production_company))
        movie_genre_dict = ast.literal_eval(str(movie_genre))
        
        movie_details['movie_production_company'] = movie_production_company_dict
        movie_details['movie_production_country'] = movie_production_country_dict
        movie_details['movie_genre'] = movie_genre_dict
        movies_all_combination_list.append(flatten(movie_details, reducer='dot'))
```

```python
movies_all_combination_list[:2]
```

```python
movie_details_flattened_df = pd.DataFrame(movies_all_combination_list)
```

# extract director name

```python
director_details = pd.read_excel('Director.xlsx')['Director']
```

```python
director_names_dict_list = []
```

```python
for director_detail in director_details:
    d = {}
    director_name = director_detail.split("'name': ")[1].split("', '")[0][1:]
    d['name'] = director_name
    director_names_dict_list.append(d)
    
```

```python
director_names_dict_list
```

```python
director_names_df = pd.DataFrame(director_names_dict_list)
```
